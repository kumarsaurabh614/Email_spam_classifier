```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('spam.csv',encoding='latin-1')
```

```python
data.head()
```

```python
data=data[['v1','v2']]
```

```python
data.head()
```

```python
random=[]                                # testing a random code for extracting the words from data
for i in data['v2'][1].split():
    random.append(i)
    
```

```python
random
```

```python
words=[]
for rows in range(data.shape[0]):
    for i in data['v2'][rows].split():
        words.append(i)
        
```

```python
words   # these are different words in all the emails
```

```python
from collections import Counter
```

```python
words_dict2=Counter(words)
```

```python
words_dict2
```

```python
len(words_dict2)
```

```python
words_dict2=words_dict2.most_common(3000)
```

```python
len(words_dict2)                 # total number of most common words are 3000 from the given data
```

```python
features=[]
for rows in range(data.shape[0]):
    letters=[]
    for i in data['v2'][rows].split():
        letters.append(i)
        letters_count=[]
    for j in words_dict2:
        letters_count.append(letters.count(j[0]))
    features.append(letters_count)
        
        
        
```

```python
features=np.array(features)
```

```python
target=[]
for i in data['v1'].values:
    if(i=='ham'):
        target.append(0)
    else:
        target.append(1)
```

```python
target=np.array(target)
```

```python
target.shape                  # checking shapes of features and target
```

```python
features.shape
```

```python
len(target)
```

```python
len(features)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

```python
classifier.fit(X_train,y_train)                 # training the machine
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)         # algorithm shows accuracy of 98 %
```
